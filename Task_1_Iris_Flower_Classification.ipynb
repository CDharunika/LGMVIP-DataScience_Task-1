**Data** **Clustering**

```python
mapping = {'Iris-setosa': 0, 'Iris-versicolor': 1, 'Iris-virginica': 2}

# Apply the mapping to the 'Species' column
dataset['species_category'] = dataset['species'].map(mapping)

# Print the resulting DataFrame
dataset.head()
```

```python
dataset.tail()
```

```python
X= dataset[['sepal_lengthCm','sepal_widthCm','petal_lengthCm','petal_widthCm']]
Y= dataset['species']
```

```python
X
```

```python
Y
```

**Test/train split**

```python
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.20,random_state=42)
```

**Training with the logistic regression model**

```python
#creatING an instance for the LogisticRegression model
lr = LogisticRegression()
lr.fit(X,Y)

# Train the model with the training set
lr.fit(X_train,Y_train)
```

```python
predictions = lr.predict(X)

# Comparing with the actual charges
Scores = pd.DataFrame({'Actual':Y,'Predictions':predictions})
Scores.head()
```

```python
#making predicions
Y_test_hat=lr.predict(X_test)
#Evolution of model
print('ACCURACY SCORE: ',accuracy_score(Y_test,Y_test_hat)*100,'%')
```

IT IS CLEAR THAT LOGISTIC REGRESSION GIVES 100% ACCURACY
